```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scx=StandardScaler()
X_train=scx.fit_transform(X_train)
X_test=scx.transform(X_test)
X_train
```

```python
scy=StandardScaler()
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
y_test
```

```python
import pickle
filename="common_datasetscx.pkl"
pickle.dump(scx,open("common_datasetscx",'wb'))
filename="common_datasetscy.pkl"
pickle.dump(scy,open("common_datasetscy",'wb'))
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=5000)
regressor=regressor.fit(X_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
y_pred=regressor.predict(X_test)
```

```python
#Evaluation
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
filename="finalized_model_svm.sav"
pickle.dump(regressor,open("finalized_model_svm",'wb'))
```

```python
preprocess_input=scx.transform([[42,43,2,0,1]])
preprocess_input
```

```python
loaded_model= pickle.load(open("finalized_model_svm",'rb'))
result=loaded_model.predict(preprocess_input)
result
```
