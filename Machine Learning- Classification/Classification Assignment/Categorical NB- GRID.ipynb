```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep=dataset[['classification_yes']]
```

```python
print(indep.iloc[0])
```

```python
dataset["classification_yes"].value_counts()
```

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import GridSearchCV
param_grid={'alpha':[1,11], 'force_alpha':[True,False], 'fit_prior':[True,False]}
grid= GridSearchCV(CategoricalNB(), param_grid, refit= True, verbose=3, n_jobs=-1, scoring= 'f1_weighted')
grid.fit(indep,dep)
```

```python
y_pred=grid.predict(indep)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(dep,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(dep,y_pred)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_score= roc_auc_score(dep,grid.predict_proba(indep)[:,1])
roc_score
```

```python
print ( " the best parameter is ", grid.best_params_)
```
